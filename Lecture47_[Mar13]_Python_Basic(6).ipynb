### `pandas.Series`

- series는 1차원 배열, dataframe은 2차원 배열 ; pandas의 대표적인 class object
- series index는 data 값과 1:1 대응, dictionary와 유사 구조


- series data type of values == int64

```
import pandas as pd
```

```
dict_data = {'a' : 1, 'b' : 2, 'c' : 3}
sr = pd.Series(dict_data)

# dictionary type to pandas.Series
print (type(sr), "\n")

print (sr, '\n')

print (sr.dtypes)
print (sr.index)
print (sr.values, '\n')
print (sr.describe)
```

```
list_data = ['2019-01-02', 3.14, 'ABC', 100, True]
sr1 = pd.Series(list_data)
idx = sr1.index
val = sr1.values

print (sr1)

print (idx)

print (val)
```

```
tup_data = ('영인', '2010-05-01', '여', True)
tup_index = ['이름', '생년월일', '성별', '학생여부']

sr2 = pd.Series( tup_data, index=tup_index,)

print (sr2)
print (sr2.index)
print ()

# 정수형 위치 인덱스
for i in range(len(sr2)) :
    print (sr2[i])
print ()
# 이름 인덱스
for j in sr2.index :
    print (sr2[j])
print()

print (sr2[['생년월일', '성별']])
print ()
print (sr2['생년월일':'성별'])
print ()
print (sr2 [1:3])
print ()
print (sr2 [[1, 2]])
```

exe) `Series sr2`로부터 인덱스 이름을 사용하여 `print(sr2[[1, 2])`와 동일한 결과를 출력하세요

```
print (sr2[[1, 2]])
print ()
print (sr2[['생년월일', '성별']])
```

# dataframe 데이터 프레임
- dataframe은 2차원 배열, R의 dataframe에서 유래
- dataframe의 열은 각각 series 객체

```
dict_data = {
    'c0' : [1,2,3],
    'c1' : [4,5,6],
    'c2' : [7,8,9],
    'c3' : [10,11,12],
    'c4' : [13,14,15]
}

df = pd.DataFrame (dict_data)

print (type (df))
print ()

print (df)
print ()
print (df.T)
```
