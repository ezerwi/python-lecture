### `thresh` : 지정 숫자 이상 null 값 있을 경우 해당 열 삭제

### `how`
 - `how=any` : NaN이 하나라도 있으면 삭제
 - `how=all` : 모두 NaN이면 삭제

```
print (titanic.columns,'\n')
titanic_thresh = titanic.dropna(axis=1, thresh=500)
print (titanic_thresh.columns,'\n')

titanic_howall = titanic.dropna (axis = 1, how = 'all')
print (titanic_howall.columns,'\n')
titanic_howany = titanic.dropna (axis = 1, how = 'any')
print (titanic_howany.columns,'\n')
```

### `fillna(method='')`
- `method='ffill'` : NaN 값을 앞의 값으로 대체
- `method='bfill'` : NaN 값을 뒤의 값으로 대체

```
titanic_ffill = titanic[:]
print (titanic['embark_town'][827:831], '\n')
titanic_ffill['embark_town'].fillna(method='ffill', inplace=True)
print (titanic_ffill['embark_town'][827:831])
```

```
titanic_bfill = titanic[:]
print (titanic['embark_town'][827:831], '\n')
titanic_bfill['embark_town'].fillna(method='bfill', inplace=True)
print (titanic_bfill['embark_town'][827:831])
```

## 중복 데이터 처리

### `duplicated` : 중복된 행 True로 반환

```
df = pd.DataFrame({
    'c1' : ['a','a','c','a','b',],
    'c2' : [1,1,1,2,2,],
    'c3' : [1,1,2,2,2,]
})

print (df)
```

```
df_dup = df.duplicated()
print (df_dup)
```

```
print (df['c2'], '\n')
print (df['c2'].duplicated().sum(), '\n')
col_dup = df['c2'].duplicated()
print (col_dup)
```

```
print (df, '\n')
print (df.drop_duplicates(), '\n')
print (df['c1'].drop_duplicates(), '\n')
print (df['c2'].drop_duplicates(), '\n')
print (df['c3'].drop_duplicates(), '\n')

print (df.drop_duplicates(subset=['c2'], ), '\n')
print (df.drop_duplicates(subset=['c2', 'c3'], ))
```

```
df3 = pd.concat([df['c1'].drop_duplicates(), df['c2'].drop_duplicates(), df['c3'].drop_duplicates()], axis=1)
print (df3)
```

# 데이터 표준화

## 데이터 단위 변경


```
auto_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/python-lecture/HW/auto-mpg.csv', header=None)
auto_df.columns = ['mpg', 'cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']
auto_df.head()
```

```
# mpg (mile per gallon) ~ kpl (kilometer per liter)

mpg_to_kpl = 1.60934 / 3.78541
print (mpg_to_kpl)
```

```
auto_df['kpl'] = auto_df.mpg * mpg_to_kpl
# auto_df.rename(columns={'mpg' : 'kpl'})
auto_df.kpl = auto_df.kpl.round(2)
display (auto_df)
```

## datatype casting

- `unique` 함수로 특이값 확인,
- `replace` 함수로 특이값 nan 처리
- `dropna` 함수로 nan 값 제거
- `astype` 함수로 type casting

```
print (auto_df.horsepower.dtype)
```

```
# unique 함수로 고유값 확인

print (auto_df.horsepower.unique())
```

```
auto_df.horsepower.replace('?', np.nan, inplace=True)
print (auto_df.horsepower.unique())
```

```
auto_df.dropna(subset=['horsepower'], axis=0, inplace=True)
print (auto_df.horsepower.unique())
```

```
print (auto_df.horsepower.isnull().sum())
print (auto_df.horsepower.dtype)
```

```
auto_df.horsepower = auto_df.horsepower.astype(float)
print (auto_df.horsepower.dtype,'\n')
print (auto_df.info())
```
