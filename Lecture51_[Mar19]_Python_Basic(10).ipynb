### `fillna(method='')`
- `method='ffill'` : NaN 값을 앞의 값으로 대체
- `method='bfill'` : NaN 값을 뒤의 값으로 대체

```
titanic_ffill = titanic[:]
print (titanic['embark_town'][827:831], '\n')
titanic_ffill['embark_town'].fillna(method='ffill', inplace=True)
print (titanic_ffill['embark_town'][827:831])
```

```
titanic_bfill = titanic[:]
print (titanic['embark_town'][827:831], '\n')
titanic_bfill['embark_town'].fillna(method='bfill', inplace=True)
print (titanic_bfill['embark_town'][827:831])
```

## `histogram`, `cut` 함수로 카테고리화

- `np.histogram`
  * `bin_counter` : `bins=3`에 따라 3개 범위로 구분
  * `count` : 범위별로 개수

- `pd.cut`
  * `x` : 참조할 데이터
  * `bins` : 경계값 리스트, 구분 기준
  * `labels` : bin label 이름
  * `include_lowest` : 첫 경계값 포함 여부

```
# print (np.histogram(auto_df.horsepower, bins=3))

count, bin_dividers = np.histogram(auto_df.horsepower, bins=3)
print (count, bin_dividers , '\n')

bin_names = ['low', 'normal','high']

auto_df['hp_bin'] = pd.cut(x=auto_df.horsepower, bins=bin_dividers, labels=bin_names, include_lowest=True)

print (auto_df[['horsepower','hp_bin']].sample(10))
```

```
hp_dummies = pd.get_dummies(auto_df['hp_bin'])
display (hp_dummies.sample(10))
```

# `sklearn.preprocessing`

- `preprocessing.LabelEncoder()`, `preprocessing.OneHotEncoder()` : 전처리를 위한 encoder 객체 생성

```
from sklearn import preprocessing
```

```
enc_sample = auto_df['hp_bin'].sample(15)

print (enc_sample, '\n')

label_encoder = preprocessing.LabelEncoder()
onehot_encoder = preprocessing.OneHotEncoder()

onehot_labeled = label_encoder.fit_transform(enc_sample)
print (onehot_labeled, '\n', type(onehot_labeled))
```

```
# onehot_reshaped = onehot_labeled.reshape(len(onehot_labeled), 1)
onehot_reshaped = onehot_labeled.reshape(-1, 1)
print (onehot_reshaped, '\n', type(onehot_reshaped))
```

# 정규화
- 각 변수의 숫자 데이터의 상대적 크기 차이 때문에 ml 분석 결과가 달라질 수 있음 ; 예를 들어, a 변수 `1~1000`, b 변수 `0~1`
- 숫자 데이터의 상대적 크기차이를 제거할 필요, 각열(변수)에 속하는 데이터 값을 동일한 크기 기준으로 나누어 정규화
- 정규화 결과 데이터의 범위는 `0~1` 또는 `-1~1`
- `각 열의 값 / 최대값` or `(각 열의 값 - 최소값)/(해당 열의 최대값 - 최소값)`


```
print (auto_df.horsepower.describe(), '\n')

#정규화
auto_df.horsepower /= abs(auto_df.horsepower.max())

print (auto_df.horsepower.describe(), '\n')
print (auto_df.horsepower.sample(15))
```

```
# 동일한 결과, 최대값 - 최소값 이용하는 방법

min_x = auto_df.horsepower - auto_df.horsepower.min()
min_max = auto_df.horsepower.max() - auto_df.horsepower.min()
auto_df.horsepower = min_x / min_max

print (auto_df.horsepower.describe(), '\n')
```
