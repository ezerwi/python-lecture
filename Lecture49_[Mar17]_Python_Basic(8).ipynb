## `reshape`
데이터 값은 동일하나, 차원이 달라 오류 발생할 수 있음, 

차원의 차수 변환하여 오류 방지

```
ar = np.arange(10)
ar1 = ar.reshape(2,5)
ar2 = ar.reshape(5,2)


print (ar, '\n')
print (ar1, '\n')
print (ar2, '\n')

print (ar.reshape(5,2,order='F'), '\n')
print (ar.reshape(5,2,order='A'), '\n')
print (ar.reshape(2,-1), '\n')
# print (ar.reshape(4,-1), '\n')
# 전체 data의 약수가 아니면 에러
print (ar.reshape(-1, 2), '\n')

print (ar2.reshape(10), '\n')
# print (ar.reshape(3,3))
# len보다 작은 행렬로 reshape할 경우 에러 발생

print ('array1 : {}차원, array2 : {}차원 array3 : {}차원'.format(array1.ndim, array2.ndim, array3.ndim))
print (f'array1 : {array1.ndim}차원, array2 : {array2.ndim}차원 array3 : {array3.ndim}차원')
```

```
list1 = [1,2,3] ; print (type(list1))
array1 = np.array(list1) ; print (type (array1))
print (array1, array1.dtype)
```

### `astype`

casting

```
array_int = np.array([1,2,3])
array_float = array_int.astype('float64')
print (array_float, array_float.dtype)

array_int1 = array_float.astype('int32')
print (array_int1, array_int1.dtype)

array_float1 = np.array([1.1, 2.1, 3.1])
array_int2 = array_float1.astype('int64')
print (array_int2, array_int2.dtype)
```

```
# exe
e = np.array([1.2,2.3,3])
print (e.astype(int), e.astype(int).dtype)
print (e.astype(float), e.astype(float).dtype)
```

## `sort`
 - `np.sort(ndarray)` : 원본 행렬 변경없이 정렬된 형태 반환

 - `ndarray.sort()` : 원본 행렬 자체를 변환


```
org_arr = np.array([3,1,9,5])
print ('원본\t: ', org_arr)
sort_arr = np.sort(org_arr)
print ('org_arr\t: ', org_arr)
print ('sort_arr\t: ', sort_arr, '\n')
org_arr.sort()
print ('org_arr\t: ', org_arr)
print ('내림차순 정렬\t: ', np.sort(org_arr)[::-1])
```

```
arr2d = np.array([[8,12],[7,1]])
print ('row 방향 정렬 \n', np.sort(arr2d, axis=0), '\n')
print ('col 방향 정렬 \n', np.sort(arr2d, axis=1))
```

```
arr3d = np.random.randint(30, size=(3,2,5))
print (arr3d, '\n')
print ('axis=0')
print (np.sort(arr3d, axis=0), '\n')
print ('axis=1')
print (np.sort(arr3d, axis=1), '\n')
```

## `argsort`

정렬된 행렬의 인덱스 반환 : 원본 행렬의 원소들에 대한 인덱스를 필요로 할 때

```
orgarr = np.array([3,1,9,5])
print (orgarr, '\n')
print (np.argsort(orgarr))
print (orgarr[np.argsort(orgarr)], '\n')
print (np.argsort(orgarr)[::-1])
print (orgarr[np.argsort(orgarr)[::-1]])
```

## `dot()` 행렬 곱(내적)

- 기본적인 행렬 곱에 따라 왼쪽 행렬 열 개수와 오른쪽 행렬 행 개수 동일해야 가능


```
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[7,8],[9,10],[11,12]])
print (np.dot(A, B))
```

```
#exe
a1 = np.arange(1,7).reshape(2,3) ; print (a1)
b1 = np.arange(10,16).reshape(3,2) ; print (b1, '\n')
print (np.dot(a1, b1))
```
