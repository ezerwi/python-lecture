#### Q1. titanic3.csv 데이터 셋을 불러와서 사용자 함수 transform_features()을 이용하여  필요한 전처리 작업을 수행한 후 아래 사항을 수행하세요. --- (30점)
- 'pclass','sex','age','fare','survived' 컬럼으로 이루어진 df1을 만들고 pickle을 이용해서 t3_df.pkl로 저장
- t3_df.pkl을 df1이름으로 불러옴

```python
import numpy as np
import pandas as pd
import seaborn as sns
import pickle
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('titanic3.csv')
display (df.sample(10))

# nulls : age, fare, cabin, embarked
def nulls (df) :
    df.age.fillna(df.age.mean(), inplace=True)
    df.fare.fillna(df.fare.mean(), inplace=True)
    df.cabin.fillna('N', inplace=True)
    df.embarked.fillna('N', inplace=True)
    return df

# drop : 'name', 'ticket', 'boat','body','home.dest'
def drops (df) :
    godrop = ['name','ticket','boat','body','home.dest']
    df.drop(godrop, axis=1, inplace=True)
    return df

# label encoding : 'cabin', 'sex', 'embarked'
def formatting (df) : 
    df['cabin'] = df['cabin'].str[:1]
    lbs = ['cabin', 'sex', 'embarked']
    for i in lbs :
        # df[i].astype('category')
        df[i] = LabelEncoder().fit(df[i]).transform(df[i])
    return df


def transform_features(df) : 
    # df = nulls(df)
    # df = drops(df)
    # df = formatting(df)
    return formatting(drops(nulls(df)))

df1 = transform_features(df)

with open ('t3_df.pkl', 'wb') as w :
    pickle.dump(df, w)
with open ('t3_df.pkl', 'rb') as r :
    df1 = pickle.load(r)

display (df1.sample(10))
```

#### Q6. 최대값 - 최소값을 반환하는 사용자 함수를 정의하고 아래사항을 수행하세요. --- (5점)

#### Q9. grouped.age 컬럼의 Z-score를 계산하는 사용자 함수를 정의하고 함수 매핑하여 age_zscore 이름으로 결과를 출력하세요. --- (10점)
