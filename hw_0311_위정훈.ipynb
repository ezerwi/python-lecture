#### 1. 다음은 학생 성적 관리 프로그램으로 리스트와 딕셔너리로 표현하고 for문을 사용해서 학생별 총점과 평균을 구했다. 딕셔너리로 객체를 각각 만드니 번거롭고 실수가 발생될 가능성이 높았다. Student 클래스를 만들고 각 학생 객체로 구성되는 학생 리스트를 작성하여 for문을 이용하여 출력하는 방식으로 프로그램을 개선 하세요.

```python
# 학생 리스트를 선언합니다.
students = [
    { "name": "조인성", "korean": 87, "math": 98, "english": 88, "science": 95 },
    { "name": "연하진", "korean": 92, "math": 98, "english": 96, "science": 98 },
    { "name": "구지연", "korean": 76, "math": 96, "english": 94, "science": 90 },
    { "name": "나선주", "korean": 98, "math": 92, "english": 96, "science": 92 },
    { "name": "윤아린", "korean": 95, "math": 98, "english": 98, "science": 98 },
    { "name": "윤명월", "korean": 64, "math": 88, "english": 92, "science": 92 }
]

# 학생을 한 명씩 반복합니다.
print("이름", "총점", "평균", sep="\t")
for student in students:
    # 점수의 총합과 평균을 구합니다.
    score_sum = student["korean"] + student["math"] +\
        student["english"] + student["science"]
    score_average = score_sum / 4
    # 출력합니다.
    print(student["name"], score_sum, score_average, sep="\t")
```

```python
class Student :
    sum, avg = 0, 0
    ss = ['korean', 'math', 'english', 'science']
    def __init__(self, stu) :
        self.stu = stu
    def make_sum(self) :
        for i in self.ss :
            self.sum += self.stu[i]
        return self.sum
    def make_avg(self) :
        return self.sum / len(self.ss)

print("이름", "총점", "평균", sep="\t")
for i in students:
    stu = Student(i)
    print (i['name'], end="\t")
    print (stu.make_sum(), end="\t")
    print (stu.make_avg())
```

#### 2. 원의 둘레와 넓이를 구하는 메소드를 포함하고 있는 Circle 클래스를 활용해서 반지름이 10인 원의 둘레와 넓이를 구하세요. 

```python
import math

class Circle:
    def __init__(self, ban):
        self.ban = ban
    def area(self) :
        return math.pi*self.ban*self.ban
    def leng(self) :
        return math.pi*self.ban*2

c = Circle(10)
print ('둘레 : ', c.leng())
print ('넓이 : ', c.area())
```

#### 5. 재귀함수를 이용해 다음 팩토리얼을 출력하세요.
1!: \

3!: \
4!: \
5!: 

```python
def facto(num) :
    if num == 1 : return num
    return num*facto(num-1)
    
print( facto (1))
print (facto (2))
print (facto (3))
print (facto (4))
print (facto (5))
```
