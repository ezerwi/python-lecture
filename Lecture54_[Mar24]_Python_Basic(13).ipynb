# Pandas 내장 그래프 도구

- 판다스는 `matplotlib` 라이브러리 기능을 일부 내장하고 있어서 별도로 import 하지 않고 간단히 그래프 그릴 수 있음
- `plot()` : 선그래프
- `plot(kind='bar')` : 막대그래프
- `plot(kind='hist')` : 히스토그램
- `plot(x='', y='', kind='scatter')` : 산점도, 상관관계 분석시 활용
- `plot(kind='box')` : 박스플롯, 특정 변수의 데이터 분포와 분산 정도에 대한 정보 제공

[`pandas.DataFrame.plot > kind=`](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html)
- `line` : line plot (default)
- `bar` : vertical bar plot
- `barh` : horizontal bar plot
- `hist` : histogram
- `box` : boxplot
- `kde` : Kernel Density Estimation plot
- `density` : same as `kde`
- `area` : area plot
- `pie` : pie plot
- `scatter` : scatter plot
- `hexbin` : hexbin plot


```
import pandas as pd

df = pd.read_excel('/content/남북한발전전력량.xlsx')

display (df.head(10))

df_ns = df.iloc[[0,5], 2:]
df_ns.index = ['SK','NK']
df_ns.columns = df_ns.columns.astype(int)
display (df_ns)
```

```
print (df_ns.plot(), '\n')
print (df_ns.T.plot())
```

```
tdf_ns = df_ns.T
print (tdf_ns.SK.head())
display (tdf_ns['SK'].head())
display (tdf_ns.loc[:, ['SK']].head())
```

```
print (tdf_ns.plot(kind='bar'))
```

histogram 은 주로 하나의 index에 대해서 분포도를 볼 때

```
print (tdf_ns.plot(kind='hist'))
```

```
print (tdf_ns['SK'].plot(kind='hist'))
```

# `matplotlib` 활용한 시각화

- `sns.set_style('{ darkgrid/whitegrid/dark/white/ticks }')`

```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic = sns.load_dataset('titanic')
sns.set_style("darkgrid")
# sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})


fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x='age', y='fare', data=titanic, ax=ax1, fit_reg=True)
sns.regplot(x='age', y='fare', data=titanic, ax=ax2, fit_reg=True)

plt.show()
```
