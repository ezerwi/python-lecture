# Time Series data 시계열 데이터

- df의 행 인덱스 사용시 시간으로 기록된 데이터를 분석하는 것이 유용
- 특정 시점 기록하는 `Timestamp`와 두 시점 사이 일정 기간 나타내는 `Period`
- `Timestamp` : 문자열 또는 숫자로 저장된 시간 데이터를 시계열 객체로 변환
  * pandas의 `to_datetime()` 사용하여 다른 자료형을 `Timestamp`를 나타내는 `datetime64`로 변환 가능

```
df = pd.read_csv('stock-data.csv')
display (df.sample(10), '\n')
print (df.info(), '\n')
```

```
df['new_Date'] = pd.to_datetime(df.Date)
print (df.new_Date.dtype, '\n')

print (df.sample(5), '\n')
print (df.info(), '\n')
```

```
df.set_index('new_Date', inplace=True)
df.drop('Date', axis=1, inplace=True)

print (df.sample(5), '\n')
print (df.info(), '\n')
```

```
dates = ['2019-01-01', '2020-03-01','2021-06-21']
ts_dates = pd.to_datetime(dates)
print (ts_dates, '\n')
print (ts_dates.values, '\n')

print (ts_dates.to_period(freq = 'D'))
print (ts_dates.to_period(freq = 'D').values, '\n')

print (ts_dates.to_period(freq = 'M'))
print (ts_dates.to_period(freq = 'A'))
```

시계열 데이터 만들기
- `date_range`
- `period_range`

```
ts_ms = pd.date_range(start='2019-01-01', end=None, periods=6, freq='MS', tz = 'Asia/Seoul')
# freq='MS' : 월초
print (ts_ms)
```

```
ts_me = pd.date_range('2019-01-01', periods=6, freq='M', tz='Asia/Seoul')
print (ts_me)

ts_3m = pd.date_range('2019/01/01',periods=6, freq='3M', tz = 'Asia/Seoul' )
print (ts_3m)
```

```
pr_m = pd.period_range(start='2019.01.01', end=None, periods=3, freq='M')
print (pr_m, '\n')

pr_h = pd.period_range(start='2019.01.01', end=None, periods=3, freq='H')
print (pr_h, '\n')

pr_2h = pd.period_range(start='2019.01.01', end=None, periods=3, freq='2H')
print (pr_2h, '\n')
```

```
import pandas as pd

def add_10(n: int) : return n + 10
def add_two_obj(ob1, ob2) : return ob1 + ob2

print (add_10(10))
print (add_two_obj(10, 10))
```

```
def mul_10 (n : int) : return n*10
def div_two_obj (a, b) : return round(a/b)

print (mul_10(20))
print (div_two_obj(100, 10))
```

```
dfage = titanic.age
print (dfage, '\n')

sr1 = dfage.apply(add_10)
print (sr1)
```

```
sr2 = dfage.apply (add_two_obj, ob2=10)
print (sr2)
```

```
sr3 = titanic.age.apply(lambda x : add_10(x))
print (sr3)
```

## `applymap()`

## titanic df 의 열 순서 바꿔 출력하기
- 열 선택 : `['pclass','sex','age','survived']`
- 열 이름의 리스트 생성
- 열 이름을 알파벳 순 정렬
- 열 이름을 기존 순서의 반대로 정렬
- 열 이름을 사용자 정의한 임의 순서로 재배치

```
titanic = sns.load_dataset('titanic')
df = titanic.loc[0:4, 'survived':'age']
# cols = list(df.columns)
cols = list(df.columns.values)
print (cols, '\n')

cols_sorted = sorted(cols)
df_sorted = df[cols_sorted]
print(df_sorted, '\n')

cols_rev = reversed(cols_sorted)
df_rev = df[cols_rev]
print (df_rev, '\n')

cols_cust = ['pclass', 'sex', 'age','survived']
df_cust = df[cols_cust]
print (df_cust)
```
