- `bar` : vertical bar plot
- `barh` : horizontal bar plot
- `hist` : histogram
- `box` : boxplot
- `kde` : Kernel Density Estimation plot
- `density` : same as `kde`
- `area` : area plot
- `pie` : pie plot
- `scatter` : scatter plot
- `hexbin` : hexbin plot

```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic = sns.load_dataset('titanic')
sns.set_style("darkgrid")
# sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})

# 그래프 객체 2개 생성
fig = plt.figure(figsize=(15,5))

# (1,2,1) :  1행 2열의 1번째
ax1 = fig.add_subplot(1,2,1)

# (1,2,2) :  1행 2열의 2번째
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x='age', y='fare', data=titanic, ax=ax1, fit_reg=True)
sns.regplot(x='age', y='fare', data=titanic, ax=ax2, fit_reg=False)

plt.show()
```

# Box plot 의미 해석

- 중앙값
- IQR (Inter Quartile Range) : 제3사분위수 - 제1사분위수
- lower whisker : Q1 - 1.5 X IQR 보다 작은 값
- upper whisker : Q3 + 1.5 X IQR 보다 큰 값
- outlier : lower whisker & upper whisker 를 벗어나는 값

```
box_df = pd.DataFrame([[25,28,29,29,30,34,35,35,37,38],[25,28,29,29,30,34,35,35,37,50]], index=['m1','m2'], columns=['y1','y2','y3','y4','y5','y6','y7','y8','y9','y10',])

display (box_df)
box_t = box_df.T
display (box_t.head())
print (box_t.describe())
```

```
box_t[['m1','m2']].plot(kind='box')
```

```
iqr = box_t.describe().loc['75%'] - box_t.describe().loc['25%']
print (iqr)
mid = box_t.describe().loc['50%']
print (mid)
lw = box_t.describe().loc['25%'] - 1.5*iqr
print (lw)
hw = box_t.describe().loc['75%'] + 1.5*iqr
print (hw)
# print (box_t.m1, type(box_t.m1))
for i in box_t.m1.tolist() : 
    if i > hw.m1 : print ('high outlier == ', i)
    if i < lw.m1 : print ('low outlier == ', i)

for i in box_t.m2.tolist() : 
    if i > hw.m2 : print ('high outlier == ', i)
    if i < lw.m2 : print ('low outlier == ', i)
```

```
df = pd.read_csv('auto-mpg.csv', header=None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','name']

def outliers(df) :
    iqr = df.describe().loc['75%'] - df.describe().loc['25%']
    print ('iqr ==', iqr)
    mid = df.describe().loc['50%']
    print ('mid ==', mid)
    lw = df.describe().loc['25%'] - 1.5*iqr
    print ('lw ==',lw)
    hw = df.describe().loc['75%'] + 1.5*iqr
    print ('hw ==', hw)

    for i in df.tolist() : 
        if i > hw : print ('high outlier == ', i)
        if i < lw : print ('low outlier == ', i)

print ('df.mpg')
outliers(df.mpg)
print ('df.cylinders')
outliers(df.cylinders)
```

```
%matplotlib inline  

import matplotlib as mpl  # 기본 설정 만지는 용도
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import matplotlib.font_manager as fm  # 폰트 관련 용도

sr_one = df_seoul.loc['경기도']

plt.plot(sr_one.index, sr_one.values, marker='o', markersize=10)

plt.style.use('ggplot')

plt.title('서울 > 경기 인구 이동', size=30)
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)

plt.xticks(rotation='vertical', size=10)
plt.legend(labels=['Seoul to Gyeonggi'], loc='best', fontsize=15)

plt.ylim(50000,800000)

# annotate arrow
plt.annotate('',
             xy=(20, 620000),    # arrow end
             xytext=(2, 290000), # arrow start
             xycoords = 'data',
             arrowprops = dict(arrowstyle='->', color='olive',lw=5)
             )

plt.annotate('',
             xy=(47, 450000),    # arrow end
             xytext=(30, 580000), # arrow start
             xycoords = 'data',
             arrowprops = dict(arrowstyle='->', color='olive',lw=5)
             )

# annotate text
plt.annotate('인구이동 증가(1970-1995)',
             xy=(10, 550000),    # arrow end
             rotation = 25,
             va = 'baseline',
             ha = 'center',
             fontsize = 15,
             )

plt.annotate('인구이동 감소(1995-2017)',
             xy=(40, 600000),    # arrow end
             rotation = -15,
             va = 'baseline',
             ha = 'center',
             fontsize = 15,
             )

plt.show()
```

## exe) 서울 -> 충남, 경북, 강원 인구 이동 그래프 출력

```
# display (df_seoul.sample(10))
col_years = list(map(str, range(1970, 2018)))

sr2 = df_seoul.loc[['충청남도','경상북도','강원도'], col_years]
# display (sr2.head(10))
plt.plot(sr2, marker='o', markersize=10)
plt.style.use('ggplot')

plt.title('Seoul -> Chungnam, Gyeongbuk, Gangwon', size=15)
plt.xlabel('Years')
plt.ylabel('Populations')

plt.xticks(rotation='vertical', size=10)

fig= plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(sr2.loc['충청남도', :], marker='o', markersize=10, markerfacecolor='green', color='olive',linewidth=2, label='Seoul-Chungnam', )
ax.plot(sr2.loc['경상북도', :], marker='o', markersize=10, markerfacecolor='blue', color='skyblue',linewidth=2, label='Seoul-Gyeongbuk')
ax.plot(sr2.loc['강원도', :], marker='o', markersize=10, markerfacecolor='red', color='magenta',linewidth=2, label='Seoul-Gangwon')

ax.set_title('서울 -> 충남, 경북, 강원', size=15, fontproperties=font_name)

ax.legend(loc='best')

ax.set_xlabel('Years', size=12)
ax.set_ylabel('Populations', size=12)
ax.set_xticklabels(col_years, rotation=90)

ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

plt.show()
```
