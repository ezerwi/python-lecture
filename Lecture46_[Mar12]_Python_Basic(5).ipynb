## `pickle`

- 객체의 형태를 유지하면서 파일에 저장하고 불러올 수 있게 하는 모듈
- binary 로 저장하게 되면 메모리가 훨씬 절감됨
- `dump()` 를 사용하여 dictionary 객체인 data를 파일에 저장


```
import pickle
with open ('test.txt', 'wb') as wb :
    data = {1: 'python', 2: 'you need'}
    pickle.dump(data, wb)

# with open ('test.txt', 'r') as r :
#     print (r.read(''))

with open ('test.txt', 'rb') as rb :
    print (pickle.load(rb))
    print ('-----')
    print (pickle.loads(rb))
```

## `time`

```
import time
print(time.time())
print(time.localtime())

# formatting
print(time.asctime( time.localtime(time.time())))

# asctime과 다르게 항상 현재 시간만
print (time.ctime())

directives = ['Z','z','H','I','M','s','d','m','Y','A','X','x', 'j']
for i in directives :
    d = '%'+i
    print (time.strftime (d, time.localtime( time.time())))
```
