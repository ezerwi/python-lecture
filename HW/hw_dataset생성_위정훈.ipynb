#### Q(응용) 'ID', 'Prod', 'Price' 3개의 column과 10개의 row으로 구성되는 데이터셋을 아래를 참조하여 생성하세요.
- ID는 1 ~ 10으로 적용
- Prod는 상품코드로서 np.random을 이용하여 10보다 적은 정수를 랜덤하게 선택
- Price는 가격으로서 np.random을 이용하여 1000 보다 적은 양의 소수를 랜덤하게 선택
- 3개의 컬럼을 결합하는 방법은 concat() 함수를 사용

### random하게 Youtube 영상 추출하기

```python
import pandas as pd
import numpy as np

vid = np.arange(3000)+1
sr_vid = pd.Series(vid, name='ID')

vgen = np.random.randint(0, 2, size = 3000)
sr_vgen = pd.Series(vgen, name = 'Gender')

vkey = np.random.randint(26, size = 3000)
sr_vkey = pd.Series(vkey, name='Keyword')

vtype = np.random.randint(10, 30, size = 3000)
sr_vtype = pd.Series(vtype, name = 'Type')

vtime = np.random.randint(0,25, size = 3000)
sr_vtime = pd.Series(vtime, name = 'Clock')

vclick = np.random.randint(20000000, size = 3000)
sr_vclick = pd.Series(vclick, name = 'clicks')

vpercent = np.random.randint(100, size = 3000)
sr_vpercent = pd.Series(vpercent, name = 'View Time Percentage of Running Time')

vgoods = np.random.randint(vclick)
sr_vgoods = pd.Series(vgoods, name = "goods")

sr_vratio = pd.Series(vgoods/vclick*100, name = 'Ratio')

ds = pd.concat ([sr_vid, sr_vgen, sr_vkey, sr_vtype, sr_vtime, sr_vpercent, sr_vclick, sr_vgoods, sr_vratio], axis = 1)
print (ds)
```
