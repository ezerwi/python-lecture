### Q. Titanic 가공한 내역을 정리하고 이를 쉽게 재사용할 후 있도록 종합 전처리 함수를 만드세요.(def transform_features(df):)

````python
titanic = sns.load_dataset('titanic')
print (titanic.info())
print (titanic.sample(10))
````

````python
%%time

from sklearn.preprocessing import LabelEncoder

titanic = sns.load_dataset('titanic')
# print (titanic.columns.tolist())
# print (titanic.describe())

def fillna(df) :
    df['age'].fillna(df.age.mean(), axis = 0, inplace=True)
    # print (df.age.unique())
    df['embark_town'].fillna(method='ffill', axis = 0, inplace=True)
    # print (df['embark_town'].unique())
    return df

def drop_features(df) :
    df.drop(['embarked', 'class', 'deck','alone','who', 'alive','adult_male'], axis=1, inplace=True)
    return df

def format_features(df) :
    formats = df.columns.tolist()
    # print (formats)
    for f in formats :
        # print(df[f].dtype)
        if df[f].dtype == 'object' : 
            df[f].astype('category')
            # print (df[f], '\n', type(df[f]))
            df[f] = LabelEncoder().fit_transform(df[f])
    return df

def transform_features(df):
    df = format_features(drop_features(fillna(df)))
    return df

titanic = transform_features(titanic)
print (titanic.sample(5),'\n')
print (titanic.info(), '\n')
````
