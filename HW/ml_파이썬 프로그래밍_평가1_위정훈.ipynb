#### 2. 아래 사용자 함수, 변수 리스트 및 2개의 내장함수를 각각 이용해서  [1, 4, 9, 16, 25], [1, 2]를 출력하세요.(내장함수 2개 사용 필수) - 10점

```python
# 함수를 선언합니다.
def power(item):
    return item * item
def under_3(item):
    return item < 3

# 변수를 선언합니다.
list_input_a = [1, 2, 3, 4, 5]
```

```python
a = []
for i in list_input_a :
    a.extend([power(i)])
print (a)

b=[]
for j in list_input_a :
    if under_3(j) :
        b.append(j)
print (b)
```

#### 5. 현재  날짜 시간을 '2020년 3월 11일 21시 53분 8초'와 같은 포멧으로 출력하세요. - 5점

```python
import time

para1 = ['Y', 'm', 'd', 'H', 'M', '-S']
para2 = ['년', '월','일','시','분','초']

para = zip(para1, para2)

for _ in para :
    x = time.strftime('%'+_[0], time.localtime(time.time()))  
    y = _[1]
    print ( f'{x}{y}' , end=" ")
```

#### 6. 현재 날짜/시간을 구하고 현재의 계절을 출력하는 프로그램을 작성하세요. - 20점
계절 기준 : 봄(3 ~ 5월), 여름(6 ~ 8월), 가을(9 ~ 11), 겨울(12 ~ 2)

```python
import time
timenow = time.asctime(time.localtime(time.time()))
print (timenow)
mon = int(time.strftime('%m', time.localtime(time.time())))
# print(int(mon))
seasons = ['봄', '여름', '가을', '겨울']
if 3<=mon<=5 : print (seasons[0])
elif 6 <= mon <= 8 : print (seasons[1])
elif 9 <= mon <= 11 : print (seasons[2])
elif 1 <= mon <= 2 or mon==12 : print (seasons[3])
```
