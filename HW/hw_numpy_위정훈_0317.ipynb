## random dataset

Youtube 투입 광고비 대비 광고효과(조회수, 좋아요수)

```python
import numpy as np
import pandas as pd

pd.options.display.float_format = '{:,.2f}'.format

vid = np.arange(3000)+1
sr_vid = pd.Series(vid, name='Video ID')

vad = np.random.randint(1, 101, size = 3000)
sr_vad = pd.Series(vad, name='Ad fee(₩1천만)')

vsubs = np.random.randint(10000000, size = 3000)
sr_vsubs = pd.Series(vsubs, name='Subscribers')

vkey = np.random.randint(1, 26, size = 3000)
sr_vkey = pd.Series(vkey, name='Keyword')

vtype = np.random.randint(10, 30, size = 3000)
sr_vtype = pd.Series(vtype, name = 'Type')

vtime = np.random.randint(0,25, size = 3000)
sr_vtime = pd.Series(vtime, name = 'UploadTime')

vwatch = np.random.randint(100, size = 3000)
sr_vwatch = pd.Series(vwatch, name = 'Avg.WatchingTimePerRunningTime')

vclick = np.random.randint(20000000, size = 3000)
sr_vclick = pd.Series(vclick, name = 'Clicks')

vgoods = np.random.randint(vclick)
sr_vgoods = pd.Series(vgoods, name = "Goods")

sr_caratio = pd.Series(vgoods/vad, name = 'Clicks/Ad Ratio')

sr_garatio = pd.Series(vclick/vad, name = 'Goods/Ad Ratio')

sr_score = pd.Series(sr_caratio*1+sr_garatio*10, name = 'Score')

ds = pd.concat ([sr_vid, sr_vad, sr_vsubs, sr_vkey, sr_vtype, sr_vtime, sr_vwatch, sr_vclick, sr_vgoods, sr_caratio, sr_garatio, sr_score], axis = 1)
print (ds)
```
